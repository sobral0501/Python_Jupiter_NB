To splice our data frame and select only specific portions of our data, there are 3 different ways of doing so.
**- .loc(),  - .iloc()  and   .ix().**

### .loc() takes 2 single/list/range operator separated by ','. The first one indicates the row and the second one indicate columns.

```python
# Return all items from the column.
df.loc[:,'CWDistance']
```

```python
# Return all items from the Multiple columns.
df.loc[:,['Age','CWDistance']]
```

```python
# Select few rows for multiple columns. Up to 9.
df.loc[:9,['Age','Height','CWDistance']]
```

```python
# Select range of rows. From 10 to 15
df.loc[10:15,['Age','Height','CWDistance']]
```

```python
# Starting from 15
df.loc[15:,['Age','Height','CWDistance']]
```

### .iloc() is integer based slicing, whereas .loc() used labels/column names.

```python
# Select the first 4 rows.
df.iloc[:4]
```

```python
# Select rows from 1 to 5 and Columns 2 to 4-1=3 ===> columns 2 and 3
df.iloc[1:5, 2:4]
```
