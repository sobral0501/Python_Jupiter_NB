```python
import pandas as pd
import matplotlib.pylab as plt
import numpy as np
url ="https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data"
df = pd.read_csv(url, header = None)

headers = ['Symboling','Normalized_losses','Make','Fuel_type','Aspiration','Num_of_doors','Body_style',
           'Drive_wheels','Engine_location','Wheel_base','Length','Width','Hight','Curb_weight','Engine_type',
           'Num_of_cylinders','Engine_size','Fuel_system','Bore','Stroke','Compression_ratio','Horsepower',
           'Peak_rpm','City_mpg','Highway_mpg','Price']

df.columns = headers
```

```python
df.head(10)
```

### Dealing with Missing Values in Python

```python
#  axis = 0  to drop the entire row
#  axis = 1  to drop the entire column
#  inplace = True  ==>  To modify the data frame, you have to set the parameter in place equal to true.

df.dropna(subset = ['Price'], axis = 0, inplace = True)
```

```python
df.head()
```

## How to replace missing values 

```python
# replace "?" to NaN

df.replace("?",np.nan, inplace=True)
df.head()
```

Identify_missing_values

<h4>Evaluating for Missing Data</h4>

The missing values are converted to default. We use the following functions to identify these missing values. There are two methods to detect missing data:

<ol>
    <li><b>.isnull()</b></li>
    <li><b>.notnull()</b></li>
</ol>
The output is a boolean value indicating whether the value that is passed into the argument is in fact missing data.

```python
missing_data = df.isnull()
missing_data.head(10)
### True = missing value   -   False = not missing value.
```

**Count missing values in each column**
<p>
Using a for loop in Python, we can quickly figure out the number of missing values in each column. As mentioned above, "True" represents a missing value, "False"  means the value is present in the dataset.  In the body of the for loop the method  ".value_counts()"  counts the number of "True" values. 

```python
for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts())
    print("")  
```

```python
### Finding the average
mean = df['Normalized_losses'].astype("float").mean(axis=0)
```

```python
print(mean)
```

```python
# Replace "NaN" by mean value in "normalized-losses" column
df['Normalized_losses'].replace(np.nan, mean, inplace = True)
```

## Data Formatting

```python
# Convert 'mpg' to L/100km in car dataset. L/100km = liter per 100 kilometers

df["City_mpg"] = 235/df["City_mpg"]
```

```python
# Renaming the Column.

df.rename(columns={"City_mpg": "City_L/100Km"}, inplace=True)
```

```python
df.head()
```

```python
# simply drop whole row with NaN in "price" column

df.dropna(subset=["Price"], axis=0, inplace=True)

# reset index, because we droped two rows
df.reset_index(drop=True, inplace=True)
```

```python
# Converting data type to proper format
df[["Normalized_losses"]] = df[["Normalized_losses"]].astype("int")

df.head()
```

## Data Normalization

* This normalization can make some statistical analyses easier down the road. By making the ranges consistent between variables, normalization enables a fair comparison between the different features, making sure they have the same impact. 

```python
### Simple Feature Scaling Method

df["Length"] = df["Length"]/df["Length"].max()
```

```python
### Min-Max Method

df["Length"] = (df["Length"] - df["Length"].min()) / (df["Length"].max() - df["Length"].min())
```

```python
###  Z-score

df["Length"] = (df["Length"] - df["Length"].mean()) / df["Length"].std()
```

##  Binning is when you group values together into bins. 

```python
bins = np.linspace(min(df['Price']), max(df['Price']),4)
```

```python
group_names = ['Low','Medium','High']
```

```python
df['Price-binned'] = pd.cut(df['Price'], bins, labels=group_names, include_lowest=True)
```
